```python
def getNearbyVenues(names, latitudes, longitudes, radius=1000):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}&day=any&time=any'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
toronto_venues = getNearbyVenues(names=toronto_df['Neighborhood'],
                                   latitudes=toronto_df['Latitude'],
                                   longitudes=toronto_df['Longitude']
                                  )
```

```python
print(toronto_venues.shape)
toronto_venues.head()
```

```python
toronto_onehot = pd.get_dummies(toronto_venues['Venue Category'],prefix = "", prefix_sep = "")
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood']
```

```python
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped
```

```python
def return_most_common_venues(row,num_of_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending = False)
    return row_categories_sorted.index.values[0:num_of_venues]
```

Create a dataframe with the 10 most common venues

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']
# create columns
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{} {} Most Common Venue'.format(ind+1,indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))
# creating the dataframe itself
neighborhood_venues_sorted = pd.DataFrame(columns = columns)
neighborhood_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhood_venues_sorted.iloc[ind,1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)
neighborhood_venues_sorted
```

```python
kclusters = 10
toronto_grouped =toronto_grouped.drop('Neighborhood', 1)
Kmeans = toronto_grouped_clustering = KMeans(n_clusters = kclusters,random_state = 0).fit(toronto_grouped)
Kmeans.labels_
```

Creating the final table

```python
neighborhood_venues_sorted.insert(0,'Cluster Label', Kmeans.labels_)

toronto_venues = df
# toronto_venues = toronto_venues.join(neighborhood_venues_sorted.set_index('Neighborhood'), on = 'Neighborhood',how = 'right')

toronto_venues = pd.merge(toronto_venues,neighborhood_venues_sorted, on = 'Neighborhood',how = 'right')
```
